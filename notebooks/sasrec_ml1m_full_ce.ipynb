```python
%%writefile /kaggle/working/sitecustomize.py
import numpy as np

aliases = {
    "float_": np.float64,
    "complex_": np.complex128,
    "unicode_": np.str_,
    "bool8": np.bool_,
}

for name, value in aliases.items():
    if not hasattr(np, name):
        setattr(np, name, value)
```
